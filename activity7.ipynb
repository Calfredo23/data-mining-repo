# Casuality Analysis
### This analysis visualize the spatial distribution of total casualties from terrorist attacks in Iraq, Syria, and Yemen during 2010–2019, for incidents with 20+ fatalities.

### INSIGHT: The Casualty Analysis reveals that Iraq, Syria, and Yemen experienced severe terrorism incidents with 20+ fatalities during 2010–2019, with the heatmap highlighting intense casualty concentrations in urban and conflict zones. These high-casualty incidents, driven by bombings, armed assaults, and suicide attacks, reflect the devastating human toll of ongoing conflicts / war, just like Iraq War (2003-2011), Syrian Civil War (2011), and Yemeni Civil War (2014).

```python
filtered_df = df[df['country_txt'].isin(['Iraq', 'Syria', 'Yemen'])]
filtered_df = filtered_df[(filtered_df['iyear'] >= 2010) & (filtered_df['iyear'] <= 2019)]
filtered_df = filtered_df[filtered_df['nkill'] >= 20]

filtered_df['total_casualties'] = filtered_df['nkill'].fillna(0) + filtered_df['nwound'].fillna(0)
```

```python
aggregated_df = filtered_df.groupby(['latitude', 'longitude'])['total_casualties'].sum().reset_index(name='casualty_count')

aggregated_df['normalized_weight'] = (aggregated_df['casualty_count'] - aggregated_df['casualty_count'].min()) / \
                                    (aggregated_df['casualty_count'].max() - aggregated_df['casualty_count'].min())

heat_data = aggregated_df[['latitude', 'longitude', 'normalized_weight']].values.tolist()

m = folium.Map(location=[33, 41], zoom_start=5)

HeatMap(heat_data).add_to(m)

m.save('casualty_heatmap_iraq_syria_yemen_2010_2019.html')
m
```

```python
max_weight_point = max(heat_data, key=lambda x: x[2])
max_weight = max_weight_point[2]

print('The max point is ', max_weight, 'in the coordinates of ', 'latitude:',max_weight_point[0],'longitude:',max_weight_point[1])
```

# Demographic Analysis
### This analysis visualizes the spatial distribution of terrorism incidents targeting the top 3 demographic groups (based on targtype1_txt like: Private Citizens & Property, Military, Government) in Syria and Yemen during 2015–2018, for incidents with 2.4 or more fatalities that are above average.

### INSIGHT: The Demographic Analysis reveals that Military, Private Citizens & Property, and Terrorists/Non-State Militia are the primary target demographics of terrorism incidents in Syria and Yemen during 2015–2018 with 10+ fatalities, as visualized in the Folium heatmap’s spatial distribution of incident density. These targeting patterns, driven by the Syrian Civil War and Yemeni Civil War, reflect strategic terrorist objectives to destabilize societies, weaken security forces, and undermine governance.

```python
filtered_df = df[df['country_txt'].isin(['Syria', 'Yemen'])]
filtered_df = filtered_df[filtered_df['iyear'].between(2015, 2018)]
filtered_df = filtered_df[filtered_df['nkill'] >= 10]

filtered_df = filtered_df.dropna(subset=['targtype1_txt'])

target_counts = filtered_df['targtype1_txt'].value_counts().reset_index()
target_counts.columns = ['Target_Type', 'Incident_Count']
target_counts['Rank'] = target_counts.index + 1  # Rank starts at 1

top_targets = target_counts.head(5)
top_targets
```

```python
filtered_df = df[df['country_txt'].isin(['Syria', 'Yemen'])]
filtered_df = filtered_df[(filtered_df['iyear'] >= 2015) & (filtered_df['iyear'] <= 2018)]
filtered_df = filtered_df[filtered_df['nkill'] >= 2.4]

# This part count incidents by target type for demographics
target_type_counts = filtered_df['targtype1_txt'].value_counts()

# This part gets the top 3 target types by incident count
top_target_types = target_type_counts.head(3).index

# This part filters for top 3 target types
filtered_df_top_targets = filtered_df[filtered_df['targtype1_txt'].isin(top_target_types)]
filtered_df_top_targets
```

```python
aggregated_df = filtered_df_top_targets.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

aggregated_df['normalized_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                    (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())


heat_data = aggregated_df[['latitude', 'longitude', 'normalized_weight']].values.tolist()

m = folium.Map(location=[33, 41], zoom_start=5)

HeatMap(heat_data).add_to(m)

m.save('demographic_heatmap_syria_yemen_2015_2018.html')
m
```

```python
max_weight_point = max(heat_data, key=lambda x: x[2])
max_weight = max_weight_point[2]

print('The max point/focus is', max_weight, 'in the coordinates of ', 'latitude:',max_weight_point[0],'longitude:',max_weight_point[1])
```
